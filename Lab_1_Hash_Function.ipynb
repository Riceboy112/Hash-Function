```python
# Nicholas Kong Assignment 2 Hash Function 

import uuid
import hashlib
 
def hash_password(password):
    # uuid is used to generate a random number
    salt = uuid.uuid4().hex
    return hashlib.sha256(salt.encode() + password.encode()).hexdigest() + ':' + salt
    
# This will take the hashed password, and compare it with each guess password that is hashed.
# So basically "Original hashed password" will be compared to see if "Bruteforce hashed password" is the same
# If the same, we cracked the hashed password. If not, we did ot crack it.
def check_password(hashed_password, user_password):
    password, salt = hashed_password.split(':')
    return password == hashlib.sha256(salt.encode() + user_password.encode()).hexdigest()


 
password = input("Please enter a password: ")
hashed_password = hash_password(password)
password_cracked = 0

print("Which database would you rather use to bruteforce the hash password?")
print("1. Database with 10,000 common passwords.")
print("2. Database with 100,000 common passwords.")
user_input = int(input("Choice: "))

# Choose between 10k or 100k common passwords
if user_input == 1:
    file_name = 'Common_Password_10k.txt'
elif user_input == 2:
    file_name = 'Common_Password_100k.txt'

with open(file_name, 'r')as file: 
    for line in file:
        for word in line.split(): 
            hacking_the_password = word
            # If statement calls to check_password
            if check_password(hashed_password, hacking_the_password):
                print("\nWe have sucessfully cracked the sha256 hashed password.")
                print("The correct password is: " + hacking_the_password)
                password_cracked = 1
                break # breaks out of the first for loop.
        else: 
            print("Trying password: " + hacking_the_password + "  does not match!")
            continue # Continues the for loop if the first nested loop did not break
        break # Breaks our of both for loops if the if statement break, breaks out both loops
                

if (password_cracked == 1):
    print("The password was sucessfully acquired!")
else:
    print("Password was not acquired! No passwords in database matched.")
```
